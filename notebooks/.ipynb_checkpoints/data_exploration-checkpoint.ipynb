```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/historic_nba_player_data.csv')
df
```

```python
df_stats = df.drop(columns=['Season'])
```

```python
df_stats.select_dtypes(include=['int64', 'float64']).corr()
```

```python
sns.heatmap(df_stats.select_dtypes(include=['int64', 'float64']).corr())
```

Notes:
- No stats share a high correlation with MVP vote share
- The most correlated stat is Value over Replacement Player (VORP)
- Some advanced stats (PER - BPM, VORP - WS) are highly correlated
  

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(df['PER'], df['Share'])
axs[0, 0].set_title('PER vs Share')
axs[0, 1].scatter(df['WS'], df['Share'])
axs[0, 1].set_title('WS vs Share')
axs[1, 0].scatter(df['BPM'], df['Share'])
axs[1, 0].set_title('BPM vs Share')
axs[1, 1].scatter(df['VORP'], df['Share'])
axs[1, 1].set_title('VORP vs Share')

fig.tight_layout()
plt.show()
```

Notes:
- PER and BPM share an inverted T shaped scatter plot
- The best seasons get around 30 PER and 10 BPM which is shown via the sharp upward line in the plot
- The continuation of the line of players with 0 vote share appear to be outliers, players who played very little minutes but performed well during that limited time
- WS and VORP have similar plots, a line of players with zero vote share with a loose positive correlation after roughly 7 WS and 3 VORP

```python
df.sort_values(by='PER', ascending=False).head(10)
```

- As noted above the highest values for PER and BPM come from players who played very little minutes. Naz Mitrou-Long played one minute and scored one three pointer, giving him inflated stats for TS%, PER, BPM and USG%
- To remove this from the training data we should cutoff players at a certain amount of MP

```python
df = df[df['MP'] >= 10]
```

```python
df.to_csv('../data/cleaned_nba_player_data.csv', index=False)
```
